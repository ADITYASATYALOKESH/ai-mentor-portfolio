```python
import os, getpass
os.environ['GEMINI_API_KEY'] = getpass.getpass('Paste your Google AI Studio key: ')
```

```python
import time
from google import genai
client = genai.Client(api_key=os.environ['GEMINI_API_KEY'])

time.sleep(5) # Add a 5-second delay before retrying the API call

resp = client.models.generate_content(
    model='gemini-2.5-flash',
    contents='Explain the 5-Layer AI Skill Pyramid in 4 short bullet points.'
)
print(resp.text)
```

```python
os.environ['GROQ_API_KEY'] = getpass.getpass('Paste your Groq API key: ')
```

```python

from openai import OpenAI
import os
client = OpenAI(
    api_key=os.environ.get("GROQ_API_KEY"),
    base_url="https://api.groq.com/openai/v1",
)

response = client.responses.create(
    input="Explain the 5-Layer AI Skill Pyramid in 4 short bullet points.",
    model="openai/gpt-oss-20b",
)
print(response.output_text)
```
